## Logistic Regression Model
- It is mainly used to predict binary classification in dataset like 0 or 1, small or large, good or bad, male or female, etc...
- It measures the relation between Independent and dependent variables by using probabilites and sigmoid function('S-shaped Curve')
- **s = 1 / (1 + exp(-values))**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10,10,20)
y = 1 / (1 + np.exp(-x))
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
## Load the dataset
df = pd.read_csv('https://raw.githubusercontent.com/nagamounika5/Datasets/master/Student_PassOrFail.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
## Seperate independent variables and target
X = df[['Self_Study_Daily','Tution_Monthly']]
Y = df['Pass_Or_Fail']
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,Y,
                                                 test_size = 0.25,
                                                 random_state = 44)
```

```python
### Apply the logistic regression model
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
```

```python
log.fit(X_train,y_train)
```

```python
Y_predict = log.predict(X_test)
Y_predict
```

```python
y_test.head(2)
```

```python
X_test.head(2)
```

```python
log.predict([[6,34]])
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,Y_predict)
```

```python
from sklearn.metrics import accuracy_score
```

```python
(117 + 123) / (117 + 123 + 10 + 0)
```

```python
accuracy_score(y_test, Y_predict)
```

## Support Vector Machine Model
- It is used to find the hyperplane which best divides the dataset into two classes
- Hyperplane is a line which linearly seperates the classes

### Maximum Width 
- w * x - b = +1
- w * x - b = 0
- w * x - b = -1
