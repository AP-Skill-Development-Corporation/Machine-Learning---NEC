```python
from sklearn.datasets import load_boston
bos_data = load_boston()
```

```python
bos_data.keys()
```

```python
df = pd.DataFrame(bos_data.data,
                  columns=bos_data.feature_names)
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
bos_data.target
```

```python
df['LandPrice'] = bos_data.target
df.head()
```

```python
# Preprocessing 
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Seperate the input and target featues
x = df.iloc[:,0:-1] # getting columns except last columns
x.head()
```

```python
y = df['LandPrice']
y
```

```python
# Apply the model
from sklearn.linear_model import LinearRegression
lin_alg = LinearRegression()
```

```python
# train the model by using fit
lin_alg.fit(x,y)
```

```python
df.shape
```

```python
# Split the data into training data and testing data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(
    x,y,train_size = 0.75,random_state = 2)
```

```python
print("LR Train Score:",lin_alg.score(x_train,y_train))
```

```python
print("LR Test Score:",lin_alg.score(x_test,y_test))
```

```python
from sklearn.linear_model import Ridge
rd = Ridge(alpha = 100)
```

```python
rd.fit(x_train,y_train)
```

```python
print('rd trained Score', rd.score(x_train,
                                   y_train))
print('rd testing Score', rd.score(x_test,
                                   y_test))
```

```python
plt.plot(lin_alg.coef_, color = 'r',
         label = 'LinearRegression')
plt.plot(rd.coef_, color = 'g', 
         label = 'RidgeRegression')
plt.title('Linear vs Ridge')
plt.xlabel('coeficient index')
plt.ylabel('coeficient magnitude')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import Ridge
rd = Ridge(alpha = 1)
```

```python
rd.fit(x_train,y_train)
```

```python
print('rd trained Score', rd.score(x_train,
                                   y_train))
print('rd testing Score', rd.score(x_test,
                                   y_test))
```

```python
plt.plot(lin_alg.coef_, color = 'r',
         label = 'LinearRegression')
plt.plot(rd.coef_, color = 'g', 
         label = 'RidgeRegression')
plt.title('Linear vs Ridge')
plt.xlabel('coeficient index')
plt.ylabel('coeficient magnitude')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import Lasso
la = Lasso(alpha = 1)
```

```python
la.fit(x_train,y_train)
```

```python
print('lasso trained Score', la.score(x_train,
                                   y_train))
print('lasso testing Score', la.score(x_test,
                                   y_test))
```

```python
plt.plot(lin_alg.coef_, color = 'r',
         label = 'LinearRegression')
plt.plot(rd.coef_, color = 'g', 
         label = 'RidgeRegression')
plt.plot(la.coef_, color = 'b', 
         label = 'LassoRegression')
plt.title('Linear vs Ridge vs Lasso')
plt.xlabel('coeficient index')
plt.ylabel('coeficient magnitude')
plt.legend()
plt.show()
```

## KNN->K-Nearest Neighbour Classifier
- knn is easy to implement 
- Support Classification and Regression
- knn algorithem assumesthat the similar things excists nearer to each other
### How does Knn works
1. select k number of neighbours,i.e k=3,5(odd values  for better result)
2. calculate the Euclidien distance b/w data points and nearest neighbours
3. Assign the new data point to the majority nearest neighbours class
